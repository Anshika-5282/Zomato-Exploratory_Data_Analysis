```python
df.columns
```

```python
df['highlights'].unique()
```

```python
all_highlights = []

for i in df['highlights'].apply(lambda x: x.replace('[','').replace(']','').replace("'",'').split(',')):
    temp_list = list(set([j.lower().strip() for j in i]))
    all_highlights.extend(temp_list)


all_highlights_unique = list(set(all_highlights))
all_highlights_unique.remove('')
print(len(all_highlights_unique))   # 470017
print(len(all_highlights))
```

```python
df['highlights'].values[0]
```

```python
case = "['Cash', 'Takeaway Available']"
case.replace('[','').replace(']','').replace("'",'').split(',')
```

```python
all_highlights_dict = {}

for i in all_highlights_unique:
    all_highlights_dict[i] = all_highlights.count(i)


highlights_df = pd.DataFrame(all_highlights_dict,index = [0]).T
highlights_df.reset_index(inplace = True)

highlights_df.columns = ['Highlights Available','Count']
```

```python
highlights_df = highlights_df.sort_values(by = 'Count',ascending = False).head(30)
```

```python
all_highlights_dict['takeaway available']
```

```python
sns.barplot(data = highlights_df,y = 'Highlights Available',x = 'Count',hue = 'Highlights Available',palette = 'magma')
```

```python
# findings 
# mostly restaurants have cash facility followed by takeway available.
# least number of restaurants provide free wi-fi and live music.
```

```python
df['highlights']
```

```python
def check_highlights(highlight_list):
    if 'Delivery' in highlight_list:
        return 'Delivery Available'
    elif 'Takeaway Available' in highlight_list:
        return 'Takeaway Available'
    else:
        return 'No Delivery Available'
```

```python
delivery_df = df['highlights'].apply(check_highlights).value_counts().reset_index()
```

```python
delivery_df
```

```python
exp = [0,0,0.3]
plt.pie(delivery_df['count'],
        labels = delivery_df['highlights'], 
        autopct = '%.1f%%',
       shadow=True,
        explode=exp)
plt.title('Restaurant Service Availability')
plt.legend(delivery_df['highlights'],loc = 1, bbox_to_anchor = [1.6,1])
plt.show()
```

```python
# findings 
# approx 44% restaurants provide takeaway facilities and 43% provide delivery facilities.
# only approx 12% restaurants provide no delivery or takeaway facility.
```
